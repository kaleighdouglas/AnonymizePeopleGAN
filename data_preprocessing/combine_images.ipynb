```python
import os
import json
import pickle
import cv2
import numpy as np
import torch

from PIL import Image
from IPython.display import display
```

```python
def process_image_directory(directory_input, directory_output):
    
    real_images = os.listdir(directory_input + '/real_B')
    if '.DS_Store' in real_images:
        real_images.remove('.DS_Store')
#     print(images)

    for image_path in real_images:
        
        ## Get Filename
        filename = image_path[:-11]
#         print(image_path)
#         print(filename)
    
        ## Read Image
        real_img = cv2.imread(directory_input + '/real_B/' + image_path)
        gen_img = cv2.imread(directory_input + '/fake_B/' + filename + '_fake_B.png')
        
        img_AB = np.concatenate([real_img, gen_img], 1)
        cv2.imwrite(directory_input + '/' + directory_output +'/'+filename+'.png', img_AB)
#         raise
```

```python
process_image_directory('combine_stage1_seed12_val_best', 'combined')
```
